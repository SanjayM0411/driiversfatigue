# **STEP 1 - Dataset Preparation and upload bold text**

```python
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)



# Check required columns
required_cols = ['Eye Blink Rate', 'Steering Variance', 'Heart Rate', 'Yawning Frequency', 'Fatigue Level']
assert all(col in df.columns for col in required_cols), "Missing required columns in uploaded dataset"
# assert df.shape[0] >= 100, "Dataset must contain at least 100 entries" # Removed this assertion
```

# **STEP 2: Import Libraries and Read Data**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

print("✅ Libraries imported successfully!")
```

## **STEP 3: Visualize the Dataset**

```python
df.head(50)
```

# **# STEP 4: Define Input and Output and split into train/**test**

```python
# Select only the first 50 rows of the DataFrame
df_50 = df.head(50)

X = df_50[required_cols[:-1]]
y = df_50['Fatigue Level']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=45)
```

# **step 5: train the model**

```python
import joblib

# Train Linear Regression model
lr = LinearRegression()
lr.fit(X_train, y_train)


# Train KNN model (assuming 'best_k' is determined elsewhere, for now I'll set a default)
# In a real scenario, you would find the best_k using cross-validation
best_k = 7

knn = KNeighborsRegressor(n_neighbors=best_k)
knn.fit(X_train, y_train)

# Save trained Linear Regression model
joblib.dump(lr, "linear_regression_model.pkl")
print("✅ Linear Regression model trained and saved as 'linear_regression_model.pkl'")

# Save trained KNN model with optimal K
joblib.dump(knn, "knn_model.pkl")
print(f"✅ KNN model (K={best_k}) trained and saved as 'knn_model.pkl'")

# Optional: show model details
print("\n📌 Model Details:")
print("Linear Regression Coefficients:", lr.coef_)
print("Best K for KNN:", best_k)
```

# **STEP 6: Predict and Plot**

```python
k_values = range(1, 30)
knn_mses = []

for k in k_values:
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred_knn = knn.predict(X_test)
    knn_mses.append(mean_squared_error(y_test, y_pred_knn))

# Plot K vs MSE
plt.figure(figsize=(10, 5))
plt.plot(k_values, knn_mses, marker='o')
plt.xlabel('K Value')
plt.ylabel('Mean Squared Error')
plt.title('Finding Optimal K Value')
plt.axvline(x=k_values[np.argmin(knn_mses)], color='r', linestyle='--',
            label=f'Best K = {k_values[np.argmin(knn_mses)]}')
plt.legend()
plt.grid()
plt.show()
```

# **Step 7: Model Comparison**

```python
# Make predictions on the test set
y_pred_lr = lr.predict(X_test)
y_pred_knn = knn.predict(X_test)

# Calculate metrics
lr_mse = mean_squared_error(y_test, y_pred_lr)
lr_r2 = r2_score(y_test, y_pred_lr)
knn_mse = mean_squared_error(y_test, y_pred_knn)
knn_r2 = r2_score(y_test, y_pred_knn)


# Metrics comparison
results = pd.DataFrame({
    'Model': ['Multiple Regression', f'KNN (K={best_k})'],
    'MSE': [lr_mse, knn_mse],
    'R² Score': [lr_r2, knn_r2]
}).set_index('Model')

print("\n🏆 FINAL COMPARISON:")
display(results.style.background_gradient(cmap='coolwarm', axis=0))

# Visual comparison
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Actual vs Predicted plot
ax1.scatter(y_test, y_pred_lr, alpha=0.5, label='Regression')
ax1.scatter(y_test, y_pred_knn, alpha=0.5, label='KNN')
ax1.plot([y.min(), y.max()], [y.min(), y.max()], 'k--')
ax1.set_xlabel('Actual Fatigue Level')
ax1.set_ylabel('Predicted Fatigue Level')
ax1.set_title('Actual vs Predicted')
ax1.legend()
ax1.grid()

# Residuals plot
residuals_lr = y_test - y_pred_lr
residuals_knn = y_test - y_pred_knn
ax2.scatter(y_pred_lr, residuals_lr, alpha=0.5, label='Regression')
ax2.scatter(y_pred_knn, residuals_knn, alpha=0.5, label='KNN')
ax2.axhline(0, color='black', linestyle='--')
ax2.set_xlabel('Predicted Values')
ax2.set_ylabel('Residuals')
ax2.set_title('Residual Analysis')
ax2.legend()
ax2.grid()

plt.tight_layout()
plt.show()
```

# **Step 8: Feature Importance**

```python
# For Regression
coef_df = pd.DataFrame({
    'Feature': required_cols[:-1],
    'Coefficient': lr.coef_
}).sort_values('Coefficient', ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x='Coefficient', y='Feature', data=coef_df, palette='viridis')
plt.title('Regression Feature Importance')
plt.xlabel('Coefficient Value')
plt.ylabel('')
plt.grid(axis='x')
plt.show()
```
